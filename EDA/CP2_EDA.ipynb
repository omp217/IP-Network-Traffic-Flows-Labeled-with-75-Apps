```python
df = pd.read_csv('data/Dataset-Unicauca-Version2-87Atts.csv', sep=',')
```

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding for 'ProtocolName'
label_encoder = LabelEncoder()
df['ProtocolName'] = label_encoder.fit_transform(df['ProtocolName'])
```

```python
# Grouping by ProtocolName and calculating the average idle time and active time for each unique protocol
grouped = df.groupby('ProtocolName')[['Idle.Mean', 'Active.Mean']].mean()

# Plotting the average idle time and active time for each unique protocol
grouped.plot(kind='bar', figsize=(15, 6))
plt.xlabel('Protocol Name')
plt.ylabel('Time (mean)')
plt.title('Average Idle Time and Active Time for Each Protocol')
plt.legend(['Idle Time (mean)', 'Active Time (mean)'])
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the mentioned columns

# Count of unique values in the 'Protocol' column
protocol_counts = df['Protocol'].value_counts()

# Plotting the count of unique values in the 'Protocol' column
plt.figure(figsize=(10, 6))
protocol_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Protocol')
plt.ylabel('Count of Unique Values')
plt.title('Count of Unique Values in Protocol Column')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the mentioned columns

# Grouping by ProtocolName and calculating the average forward packets for each unique protocol
avg_fwd_packets = df.groupby('ProtocolName')['Total.Fwd.Packets'].mean()

# Plotting the average forward packets for each unique protocol
plt.figure(figsize=(15, 8))
avg_fwd_packets.plot(kind='bar', color='orange')
plt.xlabel('Protocol Name')
plt.ylabel('Average Forward Packets')
plt.title('Average Forward Packets for Each Protocol')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python


# Grouping by ProtocolName and calculating the average backward packets for each unique protocol
avg_back_packets = df.groupby('ProtocolName')['Total.Backward.Packets'].mean()

# Plotting the average backward packets for each unique protocol
plt.figure(figsize=(10, 6))
avg_back_packets.plot(kind='bar', color='green')
plt.xlabel('Protocol Name')
plt.ylabel('Average Backward Packets')
plt.title('Average Backward Packets for Each Protocol')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Grouping by ProtocolName and calculating the average IAT mean for each unique protocol
avg_IAT_mean = df.groupby('ProtocolName')['Flow.IAT.Mean'].mean()

# Plotting the average IAT mean for each unique protocol
plt.figure(figsize=(15, 6))
avg_IAT_mean.plot(kind='bar', color='purple')
plt.xlabel('Protocol Name')
plt.ylabel('Average IAT Mean')
plt.title('Average IAT Mean for Each Protocol')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
